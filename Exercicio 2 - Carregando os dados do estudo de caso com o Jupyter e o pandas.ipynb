**Analisando as classes numéricas, é possível identificar que a empressa possui clientes com idade entre 21 e 79 anos com média de 35 anos. Iremos nos aprofundar mais para responder a questão 4**

```python
df_analise = df_limpa[['AGE','default payment next month']].copy()
bins = [21, 28, 34, 41, 79]
# As faixas etárias foram decididas de acordo com os valores descritos pelo .describe.
df_analise['AGE'] = pd.cut(df_analise['AGE'], bins, right=False)
df_analise.value_counts(df_analise['AGE'], sort=False)
```

```python
df_analise.value_counts(sort=False)
```

```python
df_analise.value_counts(sort=False, normalize=True)
```
